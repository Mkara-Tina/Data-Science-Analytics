12. UNDERSAMPLING
Build a sample dataset containing similar distribution of legit and fraudulent transactions 
Since fraud transactions are 492, take the same number of legit transactions from the 284,315 dataset and combine it with the fraud. This will give a balanced 50/50 dataset which is good for ML

```python
#this will randomly sample 492 entries from the legit dataset
legit_sample = legit.sample(n=492)
```

13. Concatenate Two Data Frames
 merge the two datasets, the 492 fraud plus the 492 sampled from legit

```python
#use axis=0 adds the data row-wise while axis=1 adds the data column wise
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
#print first 5 columns
#from the serial numbers, you can tell the datasets have been printed at random
new_dataset.head()
```

```python
#print last five columns
#from the dataset you can tell the values have been picked at random
new_dataset.tail()
```

```python
#print value count again to check if data concatenated successfully
#we have 492 of both fraud and legit transactions
new_dataset['Class'].value_counts()
```

```python
#print out the calss mean
#comparing this with the wholesome data, the mean values are almost similar
#shows that the nature of the dataset has not changed
#the 50/50 entries in both fraud and legit gives the ML a good base for training
#important step to check if we got a good or bad sample from the data
#if the sample was bad, the mean values would be very different.
new_dataset.groupby('Class').mean()
```

14. Split the Data into Features and Targets
 - Features are row headings: Time, V1, V2 etc
 - Targets are the actual values that belong to either Class 0 or 1
 - X stores the Features, Y stores the Targets

```python
# for X, drop the class Column axis=1 which represents the columns
#have not understood this part
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

15. Split Data into Training Data and Test Data

```python
#test_size of 0.2 means 2-% of data goes to Test, 80% goes to Train(randomly)
#stratify Y means there will be an even distribution of the 0 and 1 in data
#random_state to split data into two
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#X has 984 values, X_train has 787 and X_test has 197
print(X.shape, X_train.shape, X_test.shape)
```

16. Model Training by Logistic Regression
(mainly used in binary classification problems)
(its library is imported under sklearn.linear)

```python
#loading the model
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
#use fit to fit our data into the model
#check if this is working and why its not printing out bracket values of logistic regression
model.fit(X_train, Y_train)
```

17. Model Evaluation


```python
#accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
#accuracy of the training data is 94% which is a good accuracy
#when you input a value in to the model, it has a 94% chance of being correct
print('Accuracy on training data is : ', training_data_accuracy )
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
#the test data accuracy is 92.3% which is good for the model
#if the accuracies of test and training differ greatly, the ML didn't fit(find technical term for this)
#model can either be undertrained or overtrained causing underfitting
print(' Accuracy score on test data is: ', test_data_accuracy)
```
